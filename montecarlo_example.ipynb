```
import matplotlib.pyplot as plt
%matplotlib inline
#
# Stima di PI
# http://www.reactconsulting.it/react-dev-blog/posts/2016/ottobre/metodo-monte-carlo-per-il-calcolo-di-pi-greco/
#
import numpy as np
import random
 
import random
 
dots_range=5000
 
dots=0 
for tray in range(dots_range):
    x=random.random()
    y=random.random()
    plt.plot(x,y, 'b.')
    if(x**2+y**2 <= 1): 
        dots+=1
        plt.plot(x,y, 'r.')
 
pigreco=4*dots/dots_range
print("range=%d, dots=%d, Pi=%.4f" %(dots_range,dots,pigreco))
plt.show()
```
